`Algorithm1` takes in 6 parameters. A, B, C are all NumPy arrays with size A = p\*q, B = q\*r, and C = p\*r. C should be an empty array and is used to store the output.

It returns Array C = A * B

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time


def Algorithm1(A: np.ndarray, B: np.ndarray, C: np.ndarray, p: int, q: int, r: int):
    for i in range(0, p):
        for j in range(0, r):
            sum = 0
            for k in range(0, q):
                sum += A[i][k] * B[k][j]
            C[i][j] = sum
    return C
```

`Algorithm2` takes in 6 parameters. A, B, C are all NumPy arrays with size A = p\*q, B = q\*r, and C = p\*r. C should be an empty array and is used to store the output.

It returns Array C = A * B

```python
def Algorithm2(
    A: np.ndarray, B: np.ndarray, C: np.ndarray, p: int, q: int, r: int
):
    T = 5
    for I in range(0, p, T):
        for J in range(0, r, T):
            for K in range(0, q, T):
                for i in range(I, min(I + T, p)):
                    for j in range(J, min(J + T, r)):
                        sum = 0
                        for k in range(K, min(K + T, q)):
                            sum += A[i][k] * B[k][j]
                        C[i][j] = sum
    return C
```

`Algorithm3` takes in 6 parameters. A, B, C are all NumPy arrays with size A = p\*q, B = q\*r, and C = p\*r. C should be an empty array and is used to store the output.

It returns Array C = A * B

```python
def Algorithm3(
    A: np.ndarray,
    B: np.ndarray,
    C: np.ndarray,
    p: int,
    q: int,
    r: int,
    threshold: int = 64,
):  # Translated with help from ChatGPT and Github Copilot
    # Base Case
    if max(p, q, r) < threshold:
        Algorithm1(A, B, C, p, q, r)
        return C
    # Recursive Case
    if p >= q and p >= r:
        mid = p // 2
        Algorithm3(A[:mid, :], B, C[:mid, :], mid, q, r, threshold=threshold)
        Algorithm3(A[mid:, :], B, C[mid:, :], p - mid, q, r, threshold=threshold)

    elif r >= p and r >= q:
        mid = r // 2
        Algorithm3(A, B[:, :mid], C[:, :mid], p, q, mid, threshold=threshold)
        Algorithm3(A, B[:, mid:], C[:, mid:], p, q, r - mid, threshold=threshold)

    else:
        mid = q // 2

        T1 = np.empty_like(C)
        T2 = np.empty_like(C)

        Algorithm3(A[:, :mid], B[:mid, :], T1, p, mid, r, threshold=threshold)
        Algorithm3(A[:, mid:], B[mid:, :], T2, p, q - mid, r, threshold=threshold)

        C[:, :] = T1 + T2

    return C
```

`Algorithm4` takes in 6 parameters. A, B, C are all NumPy arrays. C should be an empty array and is used to store the output.

It returns Array C = A * B

```python
def Algorithm4(
    A: np.ndarray, B: np.ndarray, C: np.ndarray
):  # Translated with help from ChatGPT and Github Copilot
    n = A.shape[0]

    if n == 1:
        C[...] = Algorithm1(A, B, C, n, n, n)
        return C

    mid = n // 2

    # Split A, B, and C into 4 submatrices
    # A = [A11, A12]
    #     [A21, A22]
    A11 = A[:mid, :mid]
    A12 = A[:mid, mid:]
    A21 = A[mid:, :mid]
    A22 = A[mid:, mid:]

    # B = [B11, B12]
    #     [B21, B22]
    B11 = B[:mid, :mid]
    B12 = B[:mid, mid:]
    B21 = B[mid:, :mid]
    B22 = B[mid:, mid:]

    # C = [C11, C12]
    #     [C21, C22]
    C11 = C[:mid, :mid]
    C12 = C[:mid, mid:]
    C21 = C[mid:, :mid]
    C22 = C[mid:, mid:]

    # C11 = A11*B11 + A12*B21
    T = np.empty_like(C11)
    Algorithm4(A11, B11, C11)
    Algorithm4(A12, B21, T)
    C11 += T

    # C12 = A11*B12 + A12Â·*22
    Algorithm4(A11, B12, C12)
    Algorithm4(A12, B22, T)
    C12 += T

    # C21 = A21*B11 + A22*B21
    Algorithm4(A21, B11, C21)
    Algorithm4(A22, B21, T)
    C21 += T

    # C22 = A21*B12 + A22*B22
    Algorithm4(A21, B12, C22)
    Algorithm4(A22, B22, T)
    C22 += T

    return C
```

`Algorithm5` takes in 6 parameters. A, B, C are all NumPy arrays. C should be an empty array and is used to store the output.

It returns Array C = A * B

```python
def Algorithm5(A, B, C):  # Translated with help from ChatGPT and Github Copilot
    n = A.shape[0]
    # Base Case
    if n < 3:
        C = Algorithm1(A, B, C, n, n, n)
        return C

    mid = n // 2

    # Split A, B, and C into 4 submatrices
    # A = [A11, A12]
    #     [A21, A22]
    A11 = A[:mid, :mid]
    A12 = A[:mid, mid:]
    A21 = A[mid:, :mid]
    A22 = A[mid:, mid:]

    # B = [B11, B12]
    #     [B21, B22]
    B11 = B[:mid, :mid]
    B12 = B[:mid, mid:]
    B21 = B[mid:, :mid]
    B22 = B[mid:, mid:]

    # C = [C11, C12]
    #     [C21, C22]
    C11 = C[:mid, :mid]
    C12 = C[:mid, mid:]
    C21 = C[mid:, :mid]
    C22 = C[mid:, mid:]

    # 7 Empty Matrices
    M1 = np.empty_like(C11)
    M2 = np.empty_like(C11)
    M3 = np.empty_like(C11)
    M4 = np.empty_like(C11)
    M5 = np.empty_like(C11)
    M6 = np.empty_like(C11)
    M7 = np.empty_like(C11)

    # Perform the 7 multiplications from specification pdf
    # M1 = (A11 + A22)(B11 + B22)
    Algorithm5(A11 + A22, B11 + B22, M1)
    # M2 = (A21 + A22)B11
    Algorithm5(A21 + A22, B11, M2)
    # M3 = A11(B12 - B22)
    Algorithm5(A11, B12 - B22, M3)
    # M4 = A22(B21 - B11)
    Algorithm5(A22, B21 - B11, M4)
    # M5 = (A11 + A12)B22
    Algorithm5(A11 + A12, B22, M5)
    # M6 = (A21 - A11)(B11 + B12)
    Algorithm5(A21 - A11, B11 + B12, M6)
    # M7 = (A12 - A22)(B21 + B22)
    Algorithm5(A12 - A22, B21 + B22, M7)

    # Combine the 7 products into the 4 submatrices
    C11 = M1 + M4 - M5 + M7
    C12 = M3 + M5
    C21 = M2 + M4
    C22 = M1 - M2 + M3 + M6

    # Assign the submatrices back to C
    C[:mid, :mid] = C11
    C[:mid, mid:] = C12
    C[mid:, :mid] = C21
    C[mid:, mid:] = C22

    return C
```

## Helper Functions for Experiment 1 and 2

The following code creates 5 helper functions `runAlgorithm1, runAlgorithm2, runAlgorithm3, runAlgorithm4, runAlgorithm5` that are used to run the algorithm x amount of times.


```python
def runAlgorithm1(A, B, C, size, times):
    for _ in range(times):
        Algorithm1(A, B, C, size, size, size)
    return C


def runAlgorithm2(A, B, C, size, times):
    for _ in range(times):
        Algorithm2(A, B, C, size, size, size)
    return C


def runAlgorithm3(A, B, C, size, times):
    for _ in range(times):
        Algorithm3(A, B, C, size, size, size)
    return C


def runAlgorithm4(A, B, C, times):
    for _ in range(times):
        Algorithm4(A, B, C)
    return C


def runAlgorithm5(A, B, C, times):
    for _ in range(times):
        Algorithm5(A, B, C)
    return C
```

# Experiment 1

The Experiment1 function runs in accordance to the specifications in the assignment. It sets up by creating an empty dictionary and a list of sizes from 10-300. It creates a dataframe with the index of those sizes with the rest of the rows empty. Next, it loops through each size and creates a pair of matrices of that size, runs them through each of the algorithms, logs the results in the dataframe, and continues to the next size.

Finally, it creates a graph with sizes and times of each algorithm.

`Experiment1` takes in an input `t` for how many times it should run each algorithm. Returns nothing.

```python
def Experiment1(t):
    print("\nExperiment 1:\n")

    # Generate pairs of 30 square matrices from 10x10 to 300x300 with values from 0 to 1

    matrix_pairs = {}  # Dictionary to hold matrix pairs
    sizes = list(range(10, 301, 10))
    df = pd.DataFrame(
        index=sizes,
        columns=[
            "Size",
            "Algorithm 1",
            "Algorithm 2",
            "Algorithm 3",
            "T1(n)",
            "T2(n)",
            "T3(n)",
        ],
        dtype=float,
    )  # Dataframe to hold results

    for size in range(10, 301, 10):
        A = np.random.rand(
            size, size
        )  # generates a size x size matrix with values in [0, 1]
        B = np.random.rand(size, size)
        matrix_pairs[size] = (
            A,
            B,
        )  # Saves the matrix pair in the dictionary at index 'size'

        # Run the algorithms on the matrix pair
        C1 = np.zeros((size, size), dtype=A.dtype)
        C2 = np.zeros((size, size), dtype=A.dtype)
        C3 = np.zeros((size, size), dtype=A.dtype)

        df.loc[size] = [
            size,
            0,
            0,
            0,
            0,
            0,
            0,
        ]  # Adds a new row to the dataframe with the size and placeholders for the algorithms

        start_time = time.time()
        runAlgorithm1(A, B, C1, size, t)
        end_time = time.time()
        df.at[size, "Algorithm 1"] = (end_time - start_time) / t

        start_time = time.time()
        runAlgorithm2(A, B, C2, size, t)
        end_time = time.time()
        df.at[size, "Algorithm 2"] = (end_time - start_time) / t

        start_time = time.time()
        runAlgorithm3(A, B, C3, size, t)
        end_time = time.time()
        df.at[size, "Algorithm 3"] = (end_time - start_time) / t

        # Calculate T1(n), T2(n), T3(n)
        # T1(n) = O(n^3)
        # T2(n) = O(n^3)
        # T3(n) = O(n^3)

        df.at[size, "T1(n)"] = (size**3) / 10000000
        df.at[size, "T2(n)"] = (size**3) / 10000000  # Change scale
        df.at[size, "T3(n)"] = (size**3) / 10000000  # Change scale

    print(df)

    # matplotlib graph
    plt.plot(df["Size"], df["Algorithm 1"], label="Algorithm 1 - Empirical", marker="o")
    plt.plot(df["Size"], df["T1(n)"], label="Algorithm 1 - Theoretical", linestyle="--")
    plt.plot(df["Size"], df["Algorithm 2"], label="Algorithm 2 - Empirical", marker="s")
    plt.plot(df["Size"], df["T2(n)"], label="Algorithm 2 - Theoretical", linestyle="--")
    plt.plot(df["Size"], df["Algorithm 3"], label="Algorithm 3 - Empirical", marker="^")
    plt.plot(df["Size"], df["T3(n)"], label="Algorithm 3 - Theoretical", linestyle="--")

    plt.title("Comparison of Algorithms")
    plt.ylabel("Time (seconds)")
    plt.xlabel("Size")
    plt.legend()
    plt.grid(True)
    plt.show()


Experiment1(1)
```

`MatrixChainOrder` takes in an input of a list of ints representing the dimensions of the matrices and the number of matrices in the chain. It outputs s, which is an array s[i][j] that gives the optimal split point for multiplying at the minimum cost.

```python
def MatrixChainOrder(p, n):
    m = np.zeros((n, n), dtype=float)  # Use float for m to handle large values
    s = np.zeros((n, n), dtype=int)

    for l in range(2, n + 1):  # l is the chain length
        for i in range(1, n - l + 2):
            j = i + l - 1
            m[i - 1][j - 1] = float("inf")
            for k in range(i, j):
                q = m[i - 1][k - 1] + m[k][j - 1] + p[i - 1] * p[k] * p[j]
                if q < m[i - 1][j - 1]:
                    m[i - 1][j - 1] = q
                    s[i - 1][j - 1] = k - 1

    return s # m isn't needed for us
```

`multiplySeq` takes in a chain of matrices as the input and outputs the product of the chain multiplication

```python
def multiplySeq(chain):
    res = chain[0]
    for M in chain[1:]:
        C = np.zeros((res.shape[0], M.shape[1]), dtype=res.dtype)
        res = Algorithm1(res, M, C, res.shape[0], res.shape[1], M.shape[1])
    return res
```

`multiplyopt` takes in a chain of matrices as the input, the optimal order computed from `MatrixChainOrder`, the index of the left most matrix, and the index of the right most matrix. It outputs the product of the chain multiplication

```python
def multiplyOpt(chain, s, i, j): # Recursive function to multiply matrices, help from ChatGPT
    if i == j:
        return chain[i]
    k = s[i][j]
    left = multiplyOpt(chain, s, i, k)
    right = multiplyOpt(chain, s, k + 1, j)
    C = np.zeros((left.shape[0], right.shape[1]), dtype=left.dtype)
    return Algorithm1(left, right, C, left.shape[0], left.shape[1], right.shape[1])
```

`MatrixChainProblem` has no input and returns nothing. The function generates the random matrices and creates 19 matrix chains. It then performs chain multiplication on it using `multiplySeq` and `multiplyOpt` and times the execution of both of these functions. Finally, it saves the results to the dataframe and plots the data. 

```python
def MatrixChainProblem():
    rng = np.random.default_rng() # Random seat for reproducibility

    # Create a dataframe to hold the results
    # and fill it with zeros
    df = pd.DataFrame(
        columns=["chain_id", "upper_bound_size", "sequential_time", "optimal_time"]
    )
    # Fill dataframe with zeros
    for i in range(2, 21):
        new_row = pd.DataFrame(
            {
                "chain_id": [i],
                "upper_bound_size": [i * 10],
                "sequential_time": [0],
                "optimal_time": [0],
            }
        )
        df = pd.concat([df, new_row], ignore_index=True)

    # Generate random matrices and measure time for sequential and optimal multiplication
    # Then append results to dataframe and plot
    for j in range(2, 21):
        P = [rng.integers(10, j * 10 + 1) for _ in range(11)]
        chain = [rng.random((P[i], P[i + 1])) for i in range(10)]

        # seq
        start_time = time.time()
        seq = multiplySeq(chain)
        end_time = time.time()
        seq_time = end_time - start_time
        # optimal
        n = len(chain)
        s = MatrixChainOrder(P, n)
        start_time = time.time()
        opt = multiplyOpt(chain, s, 0, 9)
        end_time = time.time()
        opt_time = end_time - start_time
        # Append results to dataframe
        df.at[j - 2, "chain_id"] = j
        df.at[j - 2, "upper_bound_size"] = j * 10
        df.at[j - 2, "sequential_time"] = seq_time
        df.at[j - 2, "optimal_time"] = opt_time

    plt.figure()
    plt.plot(
        df["upper_bound_size"], df["sequential_time"], marker="o", label="Sequential"
    )
    plt.plot(df["upper_bound_size"], df["optimal_time"], marker="s", label="Optimal")
    plt.xlabel("Upper bound size")
    plt.ylabel("Time (seconds)")
    plt.title("Chain Multiplication Performance")
    plt.legend()
    plt.grid()
    plt.show()

    print(df)


MatrixChainProblem()
```
